# Missing Values Fill (mean/median,mode)

```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
import warnings
warnings.filterwarnings('ignore')
```

```python
df=pd.read_csv('/content/sample_product_data_with_missing.csv')
df.head()
```

```python
df.isnull().sum()
```

```python
df.shape
```

```python
# forword filling
df['Launch_Date'].fillna(method='ffill',inplace=True)
df.head()
```

```python
# backword fill
df['Rating'].fillna(method='bfill',inplace=True)
df.head()
```

```python
# mode fill
df['Product_Name'].fillna(df['Product_Name'].mode()[0],inplace=True)
df.head()
```

```python
# mean fill
df['Stock'].fillna(df['Stock'].mean(),inplace=True)
df.head()
```

```python
# median fill
df['Price'].fillna(df['Price'].median(),inplace=True)
df.head()
```

```python
# fill na by uning sklearn
from sklearn.impute import SimpleImputer


model=SimpleImputer(strategy='median')
df['Reviews']=model.fit_transform(df[['Reviews']])
df.head()
```

# Outlier Detaction


```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
import warnings
warnings.filterwarnings('ignore')
```

```python
data=pd.read_csv('/content/product_data_10k_with_gender.csv')
data.head()
```

```python
data.describe()
```

```python
sns.boxplot(data['Reviews'])
```

```python
# IQR outlire Detection ?
Q1=data['Reviews'].quantile(0.25)
Q3=data['Reviews'].quantile(0.75)
IQR=Q3-Q1
data=data[(data['Reviews']<(Q1-1.5*IQR))|(data['Reviews']>(Q3+1.5*IQR))]
sns.histplot(data['Reviews'])
```

```python
# Z-Score test
from scipy import stats
z=np.abs(stats.zscore(data['Reviews']))
data=data[(z<3)]
sns.histplot(data['Reviews'])
```

```python
# Retry generating and saving the messy dataset with 10,000 rows

import pandas as pd
import numpy as np
import random

# Function to add typos or irregular formatting
def insert_typos(text):
    if random.random() < 0.2:
        return text.lower().replace(" ", "_")
    elif random.random() < 0.3:
        return text.upper()
    else:
        return text

# Set seeds for reproducibility
np.random.seed(42)
random.seed(42)

# Generate messy dataset
product_ids = [f"P{100000 + i}" for i in range(10000)]
product_names = [insert_typos(name) for name in np.random.choice([
    "Cotton T-Shirt", "Bluetooth Speaker", "Face Wash Gel", "Leather Wallet", "Smart Watch",
    "Running Shoes", "Denim Jeans", "Power Bank", "Herbal Shampoo", "Wireless Earbuds"
], size=10000)]

categories = [insert_typos(cat) for cat in np.random.choice(
    ["Clothing", "Electronics", "Cosmetics", "Accessories", "Footwear"], size=10000)]

brands = [insert_typos(brand) for brand in np.random.choice(
    ["H&M", "boAt", "Nivea", "Wildhorn", "Noise", "Adidas", "Levis", "Mi", "Himalaya", "Realme"], size=10000)]

prices = np.random.choice([199, 499, 999, 1499, 2999, 3999, np.nan, 0], size=10000)
stocks = np.random.choice([0, 10, 25, 50, 100, 200, 500, np.nan], size=10000)
ratings = np.round(np.random.choice([3.0, 3.5, 4.0, 4.5, 5.0, np.nan], size=10000), 1)
reviews = np.random.choice([0, 10, 100, 250, 500, 1000, 2000, np.nan], size=10000)
discounts = np.random.choice([0, 10, 20, 30, 40, 50, np.nan], size=10000)
genders = np.random.choice(["Male", "Female", "Unisex", "unknown", np.nan], size=10000)
dates = pd.to_datetime(np.random.choice(pd.date_range("2020-01-01", "2024-12-31"), size=10000)).astype(str)
launch_dates = [date if random.random() > 0.15 else np.nan for date in dates]

# Create DataFrame
df_messy = pd.DataFrame({
    "Product_ID": product_ids,
    "Product_Name": product_names,
    "Category": categories,
    "Brand": brands,
    "Price": prices,
    "Stock": stocks,
    "Rating": ratings,
    "Reviews": reviews,
    "Discount (%)": discounts,
    "Gender": genders,
    "Launch_Date": launch_dates
})

# Save to CSV
csv_messy_path = "/messy_product_data_10k.csv"
df_messy.to_csv(csv_messy_path, index=False)

csv_messy_path
```
