 **SQL Commands**

```python
df = pd.read_csv('/superstore_sales_cleaned.csv')
```

```python
conn = sqlite3.connect(':memory:')  # Creates an in-memory database
df.to_sql('superstore', conn, index=False, if_exists='replace')
```

```python
# Q1 - Percentage of same-day shipping
query1 = """
SELECT
    ROUND((COUNT(DISTINCT Order_ID) / (SELECT COUNT(DISTINCT Order_ID) FROM superstore)) * 100, 2) AS Same_Day_Shipping_Percentage
FROM superstore
WHERE Order_Date = Ship_Date;
"""
print(pd.read_sql_query(query1, conn))
```

```python
# Q2 - Top 3 customers by total order value
query2 = """
SELECT Customer_Name, ROUND(SUM(Sales), 3) AS TotalOrderValue
FROM superstore
GROUP BY Customer_Name
ORDER BY TotalOrderValue DESC
LIMIT 3;
"""
print(pd.read_sql_query(query2, conn))




```

```python
# Q3 - Top 5 items with highest average sales per day
query3 = """
SELECT Product_ID, ROUND(AVG(Sales), 3) AS Average_Sales
FROM superstore
GROUP BY Product_ID
ORDER BY Average_Sales DESC
LIMIT 5;
"""
print(pd.read_sql_query(query3, conn))
```

```python
# Q4 - Average order value and ranking by customer
query4 = """
SELECT Customer_Name, ROUND(AVG(Sales), 3) AS avg_order_value,
DENSE_RANK() OVER (ORDER BY AVG(Sales) DESC) AS sales_rank
FROM superstore
GROUP BY Customer_Name;
"""
print(pd.read_sql_query(query4, conn))
```

```python
# Q5 - Customers with highest and lowest orders from each city
query5 = """
WITH cte AS (
  SELECT City, ROUND(MAX(Sales), 4) AS highest_order, ROUND(MIN(Sales), 4) AS lowest_order
  FROM superstore
  GROUP BY City
),
highest_orders AS (
  SELECT s.City, cte.highest_order, cte.lowest_order, s.Customer_Name
  FROM superstore s
  JOIN cte ON s.City = cte.City
  WHERE s.Sales = cte.highest_order
),
lowest_orders AS (
  SELECT s.City, cte.highest_order, cte.lowest_order, s.Customer_Name
  FROM superstore s
  JOIN cte ON s.City = cte.City
  WHERE s.Sales = cte.lowest_order
)
SELECT h.City, h.highest_order, h.Customer_Name AS highest_order_customer,
       l.lowest_order, l.Customer_Name AS lowest_order_customer
FROM highest_orders h
JOIN lowest_orders l ON h.City = l.City
ORDER BY h.City;
"""
print(pd.read_sql_query(query5, conn))
```

```python
# Q6 - Most demanded sub-category in West region
query6 = """
SELECT Sub_Category, ROUND(SUM(Sales), 3) AS total_quantity
FROM superstore
WHERE Region = 'West'
GROUP BY Sub_Category
ORDER BY total_quantity DESC
LIMIT 1;
"""
print(pd.read_sql_query(query6, conn))
```

```python
# Q7 - Order with highest number of items
query7 = """
SELECT Order_ID, COUNT(*) AS num_item
FROM superstore
GROUP BY Order_ID
ORDER BY num_item DESC
LIMIT 1;
"""
print(pd.read_sql_query(query7, conn))
```

```python
# Q8 - Order with highest cumulative value
query8 = """
SELECT Order_ID, ROUND(SUM(Sales), 3) AS order_value
FROM superstore
GROUP BY Order_ID
ORDER BY order_value DESC
LIMIT 1;
"""
print(pd.read_sql_query(query8, conn))
```

```python
# Q9 - Segment most likely to ship via first class
query9 = """
SELECT Segment, COUNT(Order_ID) AS num_of_ordr
FROM superstore
WHERE Ship_Mode = 'First Class'
GROUP BY Segment
ORDER BY num_of_ordr DESC;
"""
print(pd.read_sql_query(query9, conn))
```

```python
# Q10 - City with least contribution to revenue
query10 = """
SELECT City, ROUND(SUM(Sales), 3) AS TotalSales
FROM superstore
GROUP BY City
ORDER BY TotalSales ASC
LIMIT 1;
"""
print(pd.read_sql_query(query10, conn))
```

```python
# Q11 - Average shipping time
query11 = """
SELECT AVG(julianday(Ship_Date) - julianday(Order_Date)) AS avg_ship_time
FROM superstore;
"""
print(pd.read_sql_query(query11, conn))

```

```python
# Q12 - Segment with most orders and largest order in each state
query12 = """
WITH cte AS (
  SELECT State, Segment, COUNT(Order_ID) AS num_orders,
         RANK() OVER (PARTITION BY State ORDER BY COUNT(Order_ID) DESC) AS state_rank
  FROM superstore
  GROUP BY State, Segment
)
SELECT State, Segment
FROM cte
WHERE state_rank = 1;
"""
print(pd.read_sql_query(query12, conn))
```
